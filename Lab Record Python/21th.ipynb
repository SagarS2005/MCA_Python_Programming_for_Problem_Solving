```python
import pandas as pd

# 1. Load the dataset
df = pd.read_csv("sales_data.csv")

# 2. Create Total_Revenue column
df["Total_Revenue"] = df["Units_Sold"] * df["Price_Per_Unit"]

# 3. Total revenue generated for each product
revenue_per_product = df.groupby("Product")["Total_Revenue"].sum()
print("\nTotal Revenue per Product:")
print(revenue_per_product)

# 4. Total units sold for each category
units_per_category = df.groupby("Category")["Units_Sold"].sum()
print("\nTotal Units Sold per Category:")
print(units_per_category)

# 5. Average price per unit for each category
avg_price_per_category = df.groupby("Category")["Price_Per_Unit"].mean()
print("\nAverage Price per Unit for each Category:")
print(avg_price_per_category)

# 6. Product with maximum units sold
max_units_product = df.loc[df["Units_Sold"].idxmax()]
print("\nProduct with Maximum Units Sold:")
print(max_units_product)

# 7. Total revenue per month
monthly_revenue = df.groupby("Month")["Total_Revenue"].sum()
print("\nTotal Revenue per Month:")
print(monthly_revenue)

# 8. Filter products where Units_Sold > 40
high_selling_products = df[df["Units_Sold"] > 40]
print("\nProducts with Units Sold > 40:")
print(high_selling_products)

# 9. Sort dataset by Total_Revenue (descending)
sorted_df = df.sort_values(by="Total_Revenue", ascending=False)
print("\nDataset sorted by Total Revenue (Descending):")
print(sorted_df)
```
