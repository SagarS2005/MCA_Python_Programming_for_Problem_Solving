area and circumreferences of circle

```python
import math
radious=int(input("Enter radious of circle:"))
def findArea(radious):
    print("Area of circle is : ",math.pi*radious*radious)

def findCircumref(radious):
    print("Circumreference of circle is:",2*math.pi*radious)

findArea(radious)
findCircumref(radious)
```
