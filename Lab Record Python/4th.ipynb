```python
nums = input("Enter numbers separated by space: ")

# Convert to a list of integers
lst = [int(x) for x in nums.split()]
lst.sort()
print("Sorted List:", lst)

# ----- Mean -----
mean = sum(lst) / len(lst)
print("Mean:", mean)

# ----- Median -----
n = len(lst)
mid = n // 2

if n % 2 == 1:       # odd number of elements
    median = lst[mid]
else:                # even number of elements
    median = (lst[mid - 1] + lst[mid]) / 2

print("Median:", median)

# ----- Mode -----
frequency = {}  # Dictionary to count occurrences

# Count each number
for num in lst:
    frequency[num] = frequency.get(num, 0) + 1

# Find the max frequency
max_freq = max(frequency.values())

# Extract all numbers with that frequency (to handle multimodal cases)
modes = [num for num, freq in frequency.items() if freq == max_freq]

print("Mode:", modes)



```
