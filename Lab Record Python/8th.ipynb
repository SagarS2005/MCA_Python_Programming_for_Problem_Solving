```python
total_student=int(input("How many student details you want to enter:"))
name=[]
marks=[]

for i in range(0,total_student):
    name.insert(i,input("Enter name:"))
    marks.insert(i,int(input("Enter marks:")))

# tu_name=(name)
# print(tu_name)


print("name:",name)
print("marks",marks)
max=0
topper_name=''
for i in range(0,len(marks)):
    # for j in range(i+1,len(marks)-1):
    if max<marks[i]:
        max=marks[i]
        topper_name=name[i]

print("Topper name:",topper_name, "\tmarks:",max)
```
