```python
str = input("Enter String : ")
length=len(str)
vowelLength=0
def countVowel():
    global vowelLength
    for ch in str:
        if ch.lower() in 'aeiou':
            vowelLength+=1

countVowel()
str2=[""]*length
vowel=[""]*vowelLength
print("Name :",str)
print("Vowel length : " ,len(vowel))
c=0             # represent the index of vowel
for i in range(0,length):
    if str[i]=='a' or str[i]=='e' or str[i]=='i' or str[i]=='o' or str[i]=='u':
        str2[i]=' '
        vowel[c]=str[i]
        c+=1

    else:
        str2[i]=str[i]
        
print("Vowel :" ,vowel)

for i in range(0,len(str2)):
    if str2[i]==' ' :
        str2[i]=vowel[vowelLength-1]
        vowelLength-=1
print(str2)
```
