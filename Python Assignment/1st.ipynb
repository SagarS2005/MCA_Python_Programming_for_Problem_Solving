```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
rows = 4
for i in range(rows):
    print(" " * (rows - i - 1) + "*" * (2 * i + 1))
```

```python
n = 3  # controls the top half height

# Top half
for i in range(n):
    print(" " * (n - i) + "* " * (2*i + 1))

# Bottom half
for i in range(n - 2, -1, -1):
    print(" " * (n - i) + "* " * (2*i + 1))

```

```python
# Pattern D
rows = 4
cols = 3

for i in range(rows):
    for j in range(cols):
        if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
            print("*", end="   ")
        else:
            print(" ", end="   ")
    print()
```

```python
# Pattern E
n = 3
for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
