```python
filename = "file.txt"   

try:
    with open(filename, "r") as f:
        lines = f.readlines()
       
        print(lines)
        line_count = len(lines)
        word_count = 0

        for line in lines:
            word_count += len(line.split())

    print("Total Lines :", line_count)
    print("Total Words :", word_count)

except FileNotFoundError:
    print("File not found.")
```
