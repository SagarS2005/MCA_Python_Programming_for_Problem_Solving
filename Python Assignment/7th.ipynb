```python
import string

sentence = input("Enter a sentence: ")

# convert to lowercase
sentence = sentence.lower()

# remove punctuation
for p in string.punctuation:
    sentence = sentence.replace(p, "")

# split into words
words = sentence.split()

# count frequency
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

# display result
for word, count in freq.items():
    print(word, ":", count)
```
